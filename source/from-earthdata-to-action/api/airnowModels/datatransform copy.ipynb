```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from tqdm import tqdm
import pandas as pd

geolocator = Nominatim(user_agent="pollution_state_mapper")
geocode = RateLimiter(geolocator.reverse, min_delay_seconds=1)

# All pollutant DataFrames
pollutant_dfs = {
    "NO2": df_NO2,
    "CO": df_CO,
    "OZONE": df_OZONE,
    "PM10": df_PM10,
    "PM25": df_PM25,
    "SO2": df_SO2,
}

def get_state(lat, lon):
    try:
        location = geocode((lat, lon), language="en")
        if location and "address" in location.raw:
            return location.raw["address"].get("state")
    except:
        return None
    return None

for name, df in pollutant_dfs.items():
    tqdm.pandas(desc=f"Mapping states for {name}")
    df["State"] = df.progress_apply(lambda row: get_state(row["Latitude"], row["Longitude"]), axis=1)
    print(f"‚úÖ Added 'State' column to {name} with {df['State'].notna().sum()} successful lookups")
```

```python
import pandas as pd

# Assuming your pollutant DataFrames are already in this dict
pollutant_dfs = {
    "NO2": df_NO2,
    "CO": df_CO,
    "OZONE": df_OZONE,
    "PM10": df_PM10,
    "PM25": df_PM25,
    "SO2": df_SO2,
}

samples = []

for name, df in pollutant_dfs.items():
    # take up to 10 rows (handles smaller DataFrames too)
    sample = df.sample(n=min(10, len(df)), random_state=42)
    sample["Pollutant"] = name  # add a column to identify the pollutant
    samples.append(sample)

# Combine all samples
df_pollutant_sample = pd.concat(samples, ignore_index=True)

# Optional: reorder columns so Pollutant is first
cols = ["Pollutant"] + [c for c in df_pollutant_sample.columns if c != "Pollutant"]
df_pollutant_sample = df_pollutant_sample[cols]

print(df_pollutant_sample.shape)
print(df_pollutant_sample.head(15))
```

```python
# Install once (if needed):
# !pip install geopy tqdm

from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from tqdm import tqdm
import pandas as pd
import math

# --- Setup geocoder with polite rate limiting ---
geolocator = Nominatim(user_agent="pollution_state_mapper")
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)  # ~1 req/sec

# Optional: map full names -> USPS codes
STATE_NAME_TO_CODE = {
    "Alabama":"AL","Alaska":"AK","Arizona":"AZ","Arkansas":"AR","California":"CA","Colorado":"CO",
    "Connecticut":"CT","Delaware":"DE","District of Columbia":"DC","Florida":"FL","Georgia":"GA",
    "Hawaii":"HI","Idaho":"ID","Illinois":"IL","Indiana":"IN","Iowa":"IA","Kansas":"KS","Kentucky":"KY",
    "Louisiana":"LA","Maine":"ME","Maryland":"MD","Massachusetts":"MA","Michigan":"MI","Minnesota":"MN",
    "Mississippi":"MS","Missouri":"MO","Montana":"MT","Nebraska":"NE","Nevada":"NV","New Hampshire":"NH",
    "New Jersey":"NJ","New Mexico":"NM","New York":"NY","North Carolina":"NC","North Dakota":"ND",
    "Ohio":"OH","Oklahoma":"OK","Oregon":"OR","Pennsylvania":"PA","Rhode Island":"RI","South Carolina":"SC",
    "South Dakota":"SD","Tennessee":"TN","Texas":"TX","Utah":"UT","Vermont":"VT","Virginia":"VA",
    "Washington":"WA","West Virginia":"WV","Wisconsin":"WI","Wyoming":"WY"
}

# --- Small cache to avoid re-paying for repeated nearby points ---
cache = {}

def get_state_from_latlon(lat, lon, use_code=True, round_dp=3):
    """Reverse geocode to US state with caching (rounded coords)."""
    if pd.isna(lat) or pd.isna(lon):
        return None
    key = (round(lat, round_dp), round(lon, round_dp))
    if key in cache:
        return cache[key]

    try:
        loc = reverse((lat, lon), language="en", exactly_one=True)
        state_name = None
        if loc and hasattr(loc, "raw"):
            addr = loc.raw.get("address", {})
            # Nominatim keeps state in 'state' for US points
            state_name = addr.get("state")
            # Some DC cases:
            if not state_name and addr.get("city") == "Washington":
                state_name = "District of Columbia"
        if state_name and use_code:
            state_name = STATE_NAME_TO_CODE.get(state_name, state_name)
        cache[key] = state_name
        return state_name
    except Exception:
        cache[key] = None
        return None

# --- Apply to your sample dataframe ---
tqdm.pandas(desc="Reverse geocoding states")
df_pollutant_sample["State"] = df_pollutant_sample.progress_apply(
    lambda r: get_state_from_latlon(r["Latitude"], r["Longitude"], use_code=True),
    axis=1
)

print(df_pollutant_sample[["Latitude","Longitude","State"]].head())
# Optionally save:
# df_pollutant_sample.to_excel("pollutant_sample_with_state.xlsx", index=False)
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

# Example pollutant DataFrame
df = df_PM25  # replace with your dataframe

# Load US states shapefile (fast from Census)
STATES_URL = "https://www2.census.gov/geo/tiger/GENZ2018/shp/cb_2018_us_state_20m.zip"
us_states = gpd.read_file(STATES_URL).to_crs(epsg=4326)
us_states = us_states[~us_states['STUSPS'].isin(['PR','GU','VI','AS','MP'])]  # exclude territories

# Convert lat/lon ‚Üí geometry
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude), crs="EPSG:4326")

# Spatial join ‚Üí get state names
joined = gpd.sjoin(gdf, us_states[['NAME','geometry']], how='left', predicate='within')
df["State"] = joined["NAME"]

# Drop non-US rows
df = df[df["State"].notna()]

print(df[["Latitude","Longitude","State"]].head())
```
