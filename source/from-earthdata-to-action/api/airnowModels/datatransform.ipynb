```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from tqdm import tqdm
import pandas as pd

# --- setup ---
geolocator = Nominatim(user_agent="pollution_state_mapper")
geocode = RateLimiter(geolocator.reverse, min_delay_seconds=1)  # one request per second

# --- full US state list for validation ---
US_STATES = {
    "Alabama","Alaska","Arizona","Arkansas","California","Colorado","Connecticut","Delaware",
    "District of Columbia","Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa",
    "Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota",
    "Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey","New Mexico",
    "New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island",
    "South Carolina","South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington",
    "West Virginia","Wisconsin","Wyoming"
}

# --- pollutant dataframes ---
pollutant_dfs = {
    "NO2": df_NO2,
    "CO": df_CO,
    "OZONE": df_OZONE,
    "PM10": df_PM10,
    "PM25": df_PM25,
    "SO2": df_SO2,
}

# --- caching to avoid repeated lookups ---
cache = {}

def get_state(lat, lon):
    """Reverse geocode to get the full US state name."""
    if pd.isna(lat) or pd.isna(lon):
        return None
    key = (round(lat, 3), round(lon, 3))
    if key in cache:
        return cache[key]
    try:
        location = geocode((lat, lon), language="en")
        state = None
        if location and "address" in location.raw:
            state = location.raw["address"].get("state")
        cache[key] = state
        return state
    except Exception:
        cache[key] = None
        return None

# --- apply to all pollutant DataFrames ---
for name, df in pollutant_dfs.items():
    tqdm.pandas(desc=f"Mapping US states for {name}")
    df["State"] = df.progress_apply(lambda r: get_state(r["Latitude"], r["Longitude"]), axis=1)
    
    # keep only valid US states
    before = len(df)
    df.dropna(subset=["State"], inplace=True)
    df = df[df["State"].isin(US_STATES)]
    after = len(df)

    pollutant_dfs[name] = df  # reassign cleaned version
    print(f"‚úÖ {name}: {after}/{before} rows kept (US states only)")

print("\nüéâ All pollutant DataFrames now include 'State' (full name) and are US-only.")
```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

# --- your pollutant DataFrames ---
pollutant_dfs = {
    "NO2": df_NO2,
    "CO": df_CO,
    "OZONE": df_OZONE,
    "PM10": df_PM10,
    "PM25": df_PM25,
    "SO2": df_SO2,
}

# --- load official US state polygons (fast shapefile from Census) ---
STATES_URL = "https://www2.census.gov/geo/tiger/GENZ2018/shp/cb_2018_us_state_20m.zip"
us_states = gpd.read_file(STATES_URL).to_crs(epsg=4326)

# remove territories (keep only 50 states + DC)
us_states = us_states[~us_states["STUSPS"].isin(["PR", "GU", "VI", "AS", "MP"])]

# --- function to attach state names ---
def add_state_gdf(df: pd.DataFrame, name: str) -> pd.DataFrame:
    print(f"\nüåé Mapping states for {name} ...")

    # make GeoDataFrame from lat/lon
    gdf = gpd.GeoDataFrame(
        df,
        geometry=gpd.points_from_xy(df["Longitude"], df["Latitude"]),
        crs="EPSG:4326",
    )

    # spatial join to find which state polygon each point is inside
    joined = gpd.sjoin(gdf, us_states[["NAME", "geometry"]], how="left", predicate="within")

    # add state name column
    df["State"] = joined["NAME"]

    # drop rows outside US boundaries
    before = len(df)
    df = df[df["State"].notna()].copy()
    after = len(df)

    print(f"‚úÖ {name}: kept {after}/{before} rows within US states")

    # remove geometry to keep it a normal DataFrame
    df = df.drop(columns="geometry", errors="ignore")

    return df


# --- apply to all pollutant DataFrames ---
for name, df in pollutant_dfs.items():
    pollutant_dfs[name] = add_state_gdf(df, name)

```

```python
import pandas as pd
from functools import reduce

# your cleaned dfs: each has ["UTC","Value_*","AQI_*","State"]
dfs = [df_NO2_new, df_CO_new, df_OZONE_new, df_PM10_new, df_PM25_new, df_SO2_new]

def prep(df):
    # ensure datetime & aggregate duplicates (same State+UTC from multiple sites)
    out = df.copy()
    out["UTC"] = pd.to_datetime(out["UTC"], errors="coerce")
    # keep only State/UTC + pollutant columns
    keep = ["State","UTC"] + [c for c in out.columns if c.startswith(("Value_","AQI_"))]
    out = out[keep]
    # average duplicates; you can switch to .median() or .max()
    out = out.groupby(["State","UTC"], as_index=False).mean(numeric_only=True)
    return out

dfs_prep = [prep(d) for d in dfs]

# wide merge on State + UTC
df_final = reduce(
    lambda l, r: pd.merge(l, r, on=["State","UTC"], how="outer"),
    dfs_prep
).sort_values(["State","UTC"]).reset_index(drop=True)

print(df_final.shape)
print(df_final.head())

# optional: fill missing with 0 or leave NaN
# df_final = df_final.fillna(0)

# save
# df_final.to_excel("airnow_states_UTC_wide.xlsx", index=False)
# df_final.to_parquet("airnow_states_UTC_wide.parquet", index=False)
```

```python
# Visualize Value_NO2: time series, raw scatter, distribution, and optional per-state lines
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(style="whitegrid")

# Find a DataFrame that contains Value_NO2 (try df_final, df_NO2_new, df_NO2)
candidates = [
    globals().get('df_final'),
    globals().get('df_NO2_new'),
    globals().get('df_NO2')
]

df_source = None
for cand in candidates:
    if isinstance(cand, pd.DataFrame) and 'Value_NO2' in cand.columns:
        df_source = cand.copy()
        break

if df_source is None:
    raise KeyError("Could not find a DataFrame containing 'Value_NO2'. Make sure df_final, df_NO2_new, or df_NO2 is in the notebook and contains the column.")

# Ensure UTC exists and is datetime
if 'UTC' not in df_source.columns:
    raise KeyError("Selected DataFrame does not contain 'UTC' column ‚Äî cannot plot over time.")

df_source['UTC'] = pd.to_datetime(df_source['UTC'], errors='coerce')
# drop rows missing the essentials
df_source = df_source.dropna(subset=['UTC', 'Value_NO2']).copy()

# 1) Daily mean time series (helps when many points)
df_daily = df_source.set_index('UTC').resample('D')['Value_NO2'].mean()

plt.figure(figsize=(14,5))
plt.plot(df_daily.index, df_daily.values, marker='o')
plt.title('Daily mean Value_NO2')
plt.xlabel('UTC')
plt.ylabel('Value_NO2')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2) Raw scatter (sample if very large)
max_points = 20000
df_scatter = df_source.sort_values('UTC')
if len(df_scatter) > max_points:
    df_scatter = df_scatter.sample(n=max_points, random_state=42)

plt.figure(figsize=(14,4))
plt.scatter(df_scatter['UTC'], df_scatter['Value_NO2'], s=10, alpha=0.4)
plt.title('Raw Value_NO2 samples over time')
plt.xlabel('UTC')
plt.ylabel('Value_NO2')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3) Distribution + boxplot
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(df_source['Value_NO2'], bins=40, kde=True)
plt.title('Distribution of Value_NO2')
plt.subplot(1,2,2)
sns.boxplot(x=df_source['Value_NO2'])
plt.title('Boxplot of Value_NO2')
plt.tight_layout()
plt.show()

# 4) Optional: per-state mean time series for top states (if State exists)
if 'State' in df_source.columns:
    top_states = df_source['State'].value_counts().nlargest(6).index.tolist()
    df_states = df_source[df_source['State'].isin(top_states)].copy()
    # resample per state daily mean
    df_states['UTC'] = pd.to_datetime(df_states['UTC'])
    df_states = df_states.set_index('UTC')
    df_state_daily = df_states.groupby('State')['Value_NO2'].resample('D').mean().reset_index()

    plt.figure(figsize=(14,6))
    sns.lineplot(data=df_state_daily, x='UTC', y='Value_NO2', hue='State', estimator=None)
    plt.title('Daily mean Value_NO2 for top states')
    plt.xlabel('UTC')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Done ‚Äî you can save figures by uncommenting plt.savefig(...) lines if desired
```
